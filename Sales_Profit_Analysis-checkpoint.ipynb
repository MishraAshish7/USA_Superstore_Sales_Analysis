```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Original dataset
# Added encoding="lating1" because it throws an UnicodeError without it
dataset = pd.read_csv("Superstore.csv", encoding="latin1")

# Converting to dataframe
df = pd.DataFrame(dataset)
```

```python
# Converting date columns to date format
df["Order_Date"] = pd.to_datetime(df["Order_Date"], format="%d-%m-%Y")
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"], format= "%d-%m-%Y")
```

```python
# Creating some necessary columns for better analysis

# Creating month column
df["Month_Name"] = df["Ship_Date"].dt.month_name()

# Creating year column
df["Year"] = df["Ship_Date"].dt.year


#Replacing zero before calculate growth rate 
df["Profit"].replace(0, 0.001, inplace=True)
df["Sales"].replace(0.0, 0.001, inplace=True)

# Creating new column which calculates changes in sales and profit
df["Sales_Growth_Rate"] = df["Sales"].pct_change() * 100
df["Profit_Growth_Rate"] = df["Profit"].pct_change() * 100
```

```python

# Defining the order of months
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Converting Month_Name to categorical type with the specified order
df["Month_Name"] = pd.Categorical(df["Month_Name"], categories=month_order, ordered=True)
```

```python
# 1) What are the precise monthly and yearly trends in sales and profit?

# Aggregating sales and profits based on month
monthly_sales_profit = df.groupby("Month_Name").agg({"Sales": "sum", "Profit" : "sum"})

# Aggregating sales and profits based on year
yearly_sales_profit = df.groupby("Year").agg({"Sales":"sum", "Profit":"sum"})

# Visualizing Trends
fig, ((fig1, fig2), (fig3, fig4)) = plt.subplots(2,2, figsize= (16,10), sharex=False)

sns.set(style= "whitegrid", palette= "colorblind")

# Monthly trend of sales
sns.lineplot(data = monthly_sales_profit["Sales"], ls= "--", marker= "o", lw= 2, markersize= 8, color= "red", ax= fig1)
fig1.set_xlabel("Months", fontsize= 14, fontweight= "bold", color= "black")
fig1.set_ylabel("Sales ($)", fontsize= 14, fontweight= "bold", color= "black")
fig1.set_title("Monthly Sales Trend", fontsize= 16, fontweight= "bold", color= "black")
fig1.tick_params(axis= "x", rotation= 45)


# Monthly trend of profit
sns.lineplot(data = monthly_sales_profit["Profit"], ls= "--", marker= "o", lw= 2, markersize= 8, color= "blue", ax= fig2)
fig2.set_xlabel("Months", fontsize= 14, fontweight= "bold", color= "black")
fig2.set_ylabel("Profit ($)", fontsize= 14, fontweight= "bold", color= "black")
fig2.set_title("Monthly Profit Trend", fontsize= 16, fontweight= "bold", color= "black")
fig2.tick_params(axis= "x", rotation= 45)


# Yearly trend of sales
sns.barplot(x= yearly_sales_profit["Sales"].index, y= yearly_sales_profit["Sales"].values, hue= yearly_sales_profit.index, palette= "Set1", ax= fig3)
fig3.set_xlabel("Year", fontsize= 14, fontweight= "bold", color= "black")
fig3.set_ylabel("Sales ($)", fontsize= 14, fontweight= "bold", color= "black")
fig3.set_title("Yearly Sales Trend", fontsize= 16, fontweight= "bold", color= "black")
fig3.tick_params(axis= "x", rotation= 45)


# Yearly trend of profit
sns.barplot(x= yearly_sales_profit["Profit"].index, y= yearly_sales_profit["Profit"].values, hue= yearly_sales_profit.index, palette= "coolwarm", ax= fig4)
fig4.set_xlabel("Year", fontsize= 14, fontweight= "bold", color= "black")
fig4.set_ylabel("Profit ($)", fontsize= 14, fontweight= "bold", color= "black")
fig4.set_title("Yearly Profit Trend", fontsize= 16, fontweight= "bold", color= "black")
fig4.tick_params(axis= "x", rotation= 45)

# Adding borders to our visuals for better look
for ax in (fig1, fig2, fig3, fig4):
    for spine in ax.spines.values():
        spine.set_edgecolor("black")
        spine.set_linestyle("--")
        spine.set_linewidth(1.5)


plt.tight_layout()
plt.show()
```

```python
#    - What are the growth rates for sales and profit on a month-over-month and year-over-year basis?

# Aggregating average growth rate of sales and profits over month
monthly_growth_rate = df.groupby("Month_Name").agg({"Sales_Growth_Rate":"mean", "Profit_Growth_Rate":"mean"})

# Aggregating average growth rate of sales and profits over year
yearly_growth_rate = df.groupby("Year").agg({"Sales_Growth_Rate":"mean", "Profit_Growth_Rate":"mean"})

# Visulaizing 
fig, ((fig1, fig2), (fig3, fig4)) = plt.subplots(2,2, figsize= (16,10), sharex=False)

sns.set(style= "whitegrid", palette= "colorblind")

# Monthly growth rate of sales
sns.lineplot(data = monthly_growth_rate["Sales_Growth_Rate"], ls= "--", marker= "o", lw= 2, markersize= 8, color= "red", ax= fig1)
fig1.set_xlabel("Months", fontsize= 14, fontweight= "bold", color= "black")
fig1.set_ylabel("Sales Growth Rate (%)", fontsize= 14, fontweight= "bold", color= "black")
fig1.set_title("Monthly Sales Growth Rate", fontsize= 16, fontweight= "bold", color= "black")
fig1.tick_params(axis= "x", rotation= 45)


# Monthly growth rate of profit
sns.lineplot(data = monthly_growth_rate["Profit_Growth_Rate"], ls= "--", marker= "o", lw= 2, markersize= 8, color= "blue", ax= fig2)
fig2.set_xlabel("Months", fontsize= 14, fontweight= "bold", color= "black")
fig2.set_ylabel("Profit Growth Rate (%)", fontsize= 14, fontweight= "bold", color= "black")
fig2.set_title("Monthly Profit Growth Rate", fontsize= 16, fontweight= "bold", color= "black")
fig2.tick_params(axis= "x", rotation= 45)


# Yearly growth rate of sales
sns.lineplot(data = yearly_growth_rate["Sales_Growth_Rate"], ls= "--", marker= "o", lw= 2, markersize= 8, color= "green", ax = fig3)
fig3.set_xlabel("Year", fontsize= 14, fontweight= "bold", color= "black")
fig3.set_ylabel("Sales Growth Rate (%)", fontsize= 14, fontweight= "bold", color= "black")
fig3.set_title("Yearly Sales Growth Rate", fontsize= 16, fontweight= "bold", color= "black")
fig3.tick_params(axis= "x", rotation= 45)


# Yearly growth rate of profit
sns.lineplot(data = yearly_growth_rate["Profit_Growth_Rate"], ls= "--", marker= "o", lw= 2, markersize= 8, color= "orange", ax = fig4)
fig4.set_xlabel("Year", fontsize= 14, fontweight= "bold", color= "black")
fig4.set_ylabel("Profit Growth Rate (%)", fontsize= 14, fontweight= "bold", color= "black")
fig4.set_title("Yearly Profit Growth Rate", fontsize= 16, fontweight= "bold", color= "black")
fig4.tick_params(axis= "x", rotation= 45)

# Adding borders to our visuals for better look
for ax in (fig1, fig2, fig3, fig4):
    for spine in ax.spines.values():
        spine.set_edgecolor("black")
        spine.set_linestyle("-.")
        spine.set_linewidth(1.5)


plt.tight_layout()
plt.show()
```

```python
#    - Which customer segments (Consumer, Corporate) contribute most to overall sales and profit?

# Aggregating sales and profits based on customer segments
customer_seg = df.groupby("Segment").agg({"Sales":"sum", "Profit":"sum"})

# Sorting values based on most sales and customer segment
cust_seg_sales = customer_seg.sort_values(by= "Sales", ascending= False)

# Sorting values based on most profit and customer segment
cust_seg_profit = customer_seg.sort_values(by= "Profit", ascending= False)

print(f"The customer segment with most contributed sales is {cust_seg_sales["Sales"].index[0]} with Total Sales of {cust_seg_sales["Sales"].values[0]}")
print(f"The customer segment with most contributed in profit is {cust_seg_profit["Profit"].index[0]} with total profit of{cust_seg_profit["Profit"].values[0]}")
```

```python
#    - What is the distribution of sales and profit across different customer segments?

# Aggregating sales and profits based on customer segments
customer_seg = df.groupby("Segment").agg({"Sales":"sum", "Profit":"sum"})

# Visualizing 
fig, (fig1, fig2) = plt.subplots(1, 2, figsize= (14, 7))

# Define color palettes for better distinction
colors_sales = sns.color_palette('Set2', n_colors=len(customer_seg))
colors_profit = sns.color_palette('Set1', n_colors=len(customer_seg))

# Distribution of Sales based on customer segments
fig1.pie(customer_seg["Sales"], autopct="%1.1f%%", startangle= 140, wedgeprops=dict(edgecolor= "black", width= 0.3), 
         colors = colors_sales, labels= customer_seg["Sales"].index)
fig1.set_title("Distribution of Sales", fontsize= 16, fontweight= "bold", color="black")

# Distribution of Profits based on customer segments
fig2.pie(customer_seg["Profit"], autopct="%1.1f%%", startangle= 140, wedgeprops=dict(edgecolor= "black", width= 0.3), 
         colors= colors_profit, labels= customer_seg["Profit"].index)
fig2.set_title("Distribution of Profit", fontsize= 16, fontweight= "bold", color="black")

plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
#    - How do purchasing patterns (e.g., frequency, average order size) differ among customer segments?

# Agrregating based on customer segments
purchasing_patterns = df.groupby("Segment").agg({"Segment":"count", "Quantity":"mean", "Sales_Growth_Rate":"mean", "Profit_Growth_Rate":"mean"})

# Changing Column names
purchasing_patterns.columns = ["Frequency", "Avg_Qty_Order", "Avg_Sales_Growth_Rate", "Avg_Profit_Growth_Rate"]

print(purchasing_patterns)

# Visulaizing purchasing patterns separately

fig, ((fig1, fig2), (fig3, fig4)) = plt.subplots(2, 2, figsize= (16, 10), gridspec_kw={"hspace": 0.4, "wspace": 0.3})

# Number of Orders per Customer Segment
sns.barplot(x= purchasing_patterns["Frequency"].index, y= purchasing_patterns["Frequency"].values, hue= purchasing_patterns.index, palette= "crest", ax= fig1)
fig1.set_title("Number of Orders per Customer Segment", fontsize=18, fontweight="bold", color= "black")
fig1.set_xlabel("Customer Segment", fontsize=14, fontweight="bold", color= "black")
fig1.set_ylabel("Number of Orders", fontsize=14, fontweight="bold", color= "black")
fig1.tick_params(axis="x", rotation=45)
fig1.grid(axis= "y", ls= "--", alpha=0.7)

# Average Quantity per Order per Customer Segment
sns.barplot(x= purchasing_patterns["Avg_Qty_Order"].index, y= purchasing_patterns["Avg_Qty_Order"].values, hue= purchasing_patterns.index, palette= "cubehelix", ax= fig2)
fig2.set_title("Average Quantity per Order per Customer Segment", fontsize=18, fontweight="bold", color= "black")
fig2.set_xlabel("Customer Segment", fontsize=14, fontweight="bold", color= "black")
fig2.set_ylabel("Average Quantity per Order", fontsize=14, fontweight="bold", color= "black")
fig2.tick_params(axis="x", rotation=45)
fig2.grid(axis= "y", ls= "--", alpha=0.7)

# Average Sales Growth Rate per Customer Segment
sns.barplot(x= purchasing_patterns["Avg_Sales_Growth_Rate"].index, y= purchasing_patterns["Avg_Sales_Growth_Rate"].values, hue= purchasing_patterns.index, palette= "rocket", ax= fig3)
fig3.set_title("Average Sales Growth Rate per Customer Segment", fontsize=18, fontweight="bold", color= "black")
fig3.set_xlabel("Customer Segment", fontsize=14, fontweight="bold", color= "black")
fig3.set_ylabel("Average Sales Growth Rate (%)", fontsize=14, fontweight="bold", color= "black")
fig3.tick_params(axis="x", rotation=45)
fig3.grid(axis= "y", ls= "--", alpha=0.7)

# Average Profit growth Rate per Customer Segment
sns.barplot(x= purchasing_patterns["Avg_Profit_Growth_Rate"].index, y= purchasing_patterns["Avg_Profit_Growth_Rate"].values, hue= purchasing_patterns.index, palette= "magma", ax= fig4)
fig4.set_title("Average Profit growth Rate per Customer Segment", fontsize=18, fontweight="bold", color= "black")
fig4.set_xlabel("Customer Segment", fontsize=14, fontweight="bold", color= "black")
fig4.set_ylabel("Average Profit growth Rate (%)", fontsize=14, fontweight="bold", color= "black")
fig4.tick_params(axis="x", rotation=45)
fig4.grid(axis= "y", ls= "--", alpha=0.7)

# Adding borders to our visuals for better look
for ax in (fig1, fig2, fig3, fig4):
    for spine in ax.spines.values():
        spine.set_edgecolor("black")
        spine.set_linestyle("--")
        spine.set_linewidth(1.5)

# Manual adjustments to layout
fig.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9, hspace=0.4, wspace=0.3)
plt.show()
```

```python
#    - What are the top 10 products and product categories by total sales and profit?

# Creating a new column which will indicate the first 5 characters of Product Names because the names is too big
df["Pcode"] = df["Product_Name"].str[:5]

# Aggregating and sorting top 10 products by total sales and profits
by_product_name = df.groupby("Pcode").agg({"Sales":"sum", "Profit":"sum"}).sort_values(by= "Sales", ascending= False).head(10)

# Aggregating and sorting top products category by total sales and profits based on category
by_category = df.groupby("Category").agg({"Sales":"sum", "Profit":"sum"}).sort_values(by= "Sales", ascending= False).reset_index()

# Melting the DataFrame to have 'variable' and 'value' columns for easier plotting
melted_by_category = by_category.melt(id_vars= "Category", value_vars= ["Sales", "Profit"],
                                     var_name= "Metric", value_name= "Amount")

# Visualizing
fig, (fig1, fig2) = plt.subplots(1, 2, figsize= (16,10))

# Visualizing Top 10 Products based on total sales and profit
sns.lineplot(x= by_product_name["Sales"].index, y= by_product_name["Sales"].values, ls= "--", lw= 2, marker= "o",
             markersize= 8, label= "Sales", color= "red", ax= fig1)
sns.lineplot(x= by_product_name["Profit"].index, y= by_product_name["Profit"].values, ls= "-.", lw= 2, marker= "^",
             markersize= 8, label= "Profit", color= "blue", ax= fig1)
fig1.set_xlabel("Product Name", fontsize= 14, fontweight= "bold", color= "black")
fig1.set_ylabel("Amounts ($)", fontsize= 14, fontweight= "bold", color= "black")
fig1.set_title("Top 10 Products based on total sales and profit", fontsize= 16, fontweight= "bold", color= "black")
fig1.tick_params(axis= "x", rotation= 45)
fig1.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

# Visualizing Top Product Category by Total Sales and Profit
sns.barplot(x= melted_by_category["Category"], y= melted_by_category["Amount"], hue= "Metric", palette= "viridis", data= melted_by_category, ax= fig2)
fig2.set_xlabel("Product Category", fontsize= 14, fontweight= "bold", color= "black")
fig2.set_ylabel("Amount ($)", fontsize= 14, fontweight= "bold", color= "black")
fig2.set_title("Top Product Category by Total Sales and Profit", fontsize= 16, fontweight= "bold", color= "black")
fig2.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

# Adding data labels
for p in fig2.patches:
    fig2.annotate(f'{p.get_height():,.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', xytext=(0, 5), textcoords='offset points',
                  fontsize=10, color='black')

# Adding borders to our visual
for ax in (fig1, fig2):
    for spine in ax.spines.values():
        spine.set_edgecolor("black")
        spine.set_linewidth(1.5)

plt.tight_layout()
plt.show()
```

```python
#    - Which products and categories have the highest and lowest profit margins?

# Creating new column which will show profit margins
df["Profit_Margin"] = ( df["Profit"] / df["Sales"]) * 100

# Aggregating and sorting to find product category with highest and lowest average profit margins
pm_by_category = df.groupby("Category")["Profit_Margin"].mean().sort_values(ascending= False)

# Aggregating and sorting to find products with highest and lowest average profit margins
pm_by_Product = df.groupby("Product_Name")["Profit_Margin"].mean().sort_values(ascending= False)

print(f"The highest average profit margin is for the category '{pm_by_category.index[0]}', with a value of {pm_by_category.values[0]:.2f}%.")
print(f"The lowest average profit margin is for the category '{pm_by_category.index[-1]}', with a value of {pm_by_category.values[-1]:.2f}%.")
print(f"The highest average profit margin is for the Product Name '{pm_by_Product.index[0]}', with a value of {pm_by_Product.values[0]:.2f}%.")
print(f"The highest average profit margin is for the Product Name '{pm_by_Product.index[-1]}', with a value of {pm_by_Product.values[-1]:.2f}%.")

#
```

```python
#    - How do discounts affect sales volume and profitability for different products?

to_corr = df[["Discount", "Quantity", "Sales", "Profit", "Profit_Margin"]]

# Plotting a heatmap to visualize correlation
plt.figure(figsize=(8,6))

sns.heatmap(to_corr.corr(), annot= True, cmap= "viridis", vmin= -1, vmax= 1, center= 0, square= True, linewidths= .5, 
            cbar_kws={"shrink": .8}, annot_kws={"size": 12, "weight": "bold"})
plt.title("Correlation Between Discount, Quantity, Sales volume, Profit and Profit Margin",fontsize=16, fontweight="bold", color="black", pad=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#    - How do different discount levels impact sales volume and overall profitability?


# Aggregating by discount levels
by_discount_levels = df.groupby("Discount").agg({"Sales": "sum", "Profit": "sum"}).reset_index()
by_discount_levels.columns = ["Discount", "Total_Sales", "Total_Profit"]

# Melting for easier plotting
by_discount_levels_melted = by_discount_levels.melt(id_vars="Discount", value_vars=["Total_Sales", "Total_Profit"],
                                                   var_name="Metric", value_name="Amount")

# Plotting
plt.figure(figsize=(14, 8))

# Barplot with hue for different metrics
sns.barplot(x="Discount", y="Amount", hue="Metric", data=by_discount_levels_melted, palette="viridis")

# Adding data labels on the bars
for patch in plt.gca().patches:
    plt.text(
        patch.get_x() + patch.get_width() / 2, patch.get_height() + 0.02 * patch.get_height(),
        f'{patch.get_height():,.0f}', ha='center', va='bottom',
        fontsize=10, color='black', weight='bold'
    )

# Customizing plot title and labels
plt.title("Impact of Discount Levels on Sales Volume and Overall Profitability", fontsize=16, fontweight="bold", color="black")
plt.xlabel("Discount Levels (%)", fontsize=14, fontweight="bold", color="black")
plt.ylabel("Amount ($)", fontsize=14, fontweight="bold", color="black")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="Metric", title_fontsize='13', fontsize='12')

# Ensuring layout is tight
plt.tight_layout()
plt.show()
```

```python
#    - Are there patterns or thresholds in discounting that maximize profit?

# Aggregating by discount levels to find average profit
discount_analysis = df.groupby("Discount").agg({"Profit": "mean", "Sales": "sum"}).reset_index()
discount_analysis.columns = ["Discount", "Average_Profit", "Total_Sales"]

# Visualizing the relationship between discount and average profit
plt.figure(figsize=(14, 8))

# Plot Average Profit vs Discount Levels
sns.lineplot(x="Discount", y="Average_Profit", data=discount_analysis, marker='o', color='b', label='Average Profit')

# Adding a secondary axis for total sales
ax2 = plt.gca().twinx()
sns.lineplot(x="Discount", y="Total_Sales", data=discount_analysis, marker='o', color='r', ax=ax2, label='Total Sales')

# Adding data labels for profit
for i, row in discount_analysis.iterrows():
    plt.text(row["Discount"], row["Average_Profit"], f'{row["Average_Profit"]:.2f}', fontsize=10, color='blue', ha='center')
    
# Adding data labels for sales on secondary axis
for i, row in discount_analysis.iterrows():
    ax2.text(row["Discount"], row["Total_Sales"], f'{row["Total_Sales"]:.2f}', fontsize=10, color='red', ha='center')

# Customizing the plot
plt.title("Impact of Discount Levels on Average Profit and Total Sales", fontsize=16, fontweight="bold", color="black")
plt.xlabel("Discount Levels (%)", fontsize=14, fontweight="bold", color="black")
plt.ylabel("Average Profit ($)", fontsize=14, fontweight="bold", color="black")
ax2.set_ylabel("Total Sales ($)", fontsize=14, fontweight="bold", color="black")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax2.tick_params(axis='y', labelsize=12)

# Adding legends
plt.legend(loc='upper left')
ax2.legend(loc='upper right')

# Ensuring layout is tight
plt.tight_layout()
plt.show()
```
